# **Rock vs Mine Prediction **

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset[60].value_counts()
```

```python
dataset.groupby(60).mean()
```

# Seperting data and label

```python
X = dataset.drop(columns=60,axis=1)
Y = dataset[60]
```

Splitting Training and Test Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=1,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training --> LogisticRegression Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

# Checking Accuracy of the Model

```python
#checking accuracy for train data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on traaining data is: ',training_data_accuracy)
```

```python
#checking accuracy for test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy socre of test data is: ',test_data_accuracy)
```
